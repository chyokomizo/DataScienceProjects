##**Exploratory Data Analysis (EDA)**

```
#importing Seaborn library
import seaborn as sns
```

###**Box plot**

```
sns.boxplot(x='price', y='drive-wheels', data=df)
```

```
import matplotlib
import matplotlib.pyplot as plt
```

###**Scatter Plot**

```
#Scatter plot using matplotlib.pyplot lib
y =df['engine-size']
x = df['price']
matplotlib.pyplot.scatter(x,y,)
plt.title('Price vs. Engine-size')
plt.xlabel('Price')
plt.ylabel('Engine-size')
```

```
df_test = df[['drive-wheels','body-style', 'price']]
```

```
df_grp = df_test.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
```

```
df_grp
```

```
#Using the pivot() method to show the data in rows and columns
df_pivot = df_grp.pivot(index='drive-wheels', columns='body-style')
df_pivot
```

###**Heat Map**

```
#Using HEATMAP function
plt.pcolor(df_pivot, cmap='GnBu') #apply the function defining the dataframe tha will be used (df_pivot) and the color scheme (Green/Blue)
plt.colorbar()
plt.xlabel('body-style')
plt.ylabel('drive-whels')
plt.show()
```

```
#Analysis of Variance (ANOVA Method) 
df_anova = df[['make', 'price']]
grouped_anova = df_anova.groupby(['make'])
```

```
import statistics
from scipy import stats
```

```
#ANOVA test between Honda and Subaru cars
anova_results_1 = stats.f_oneway(grouped_anova.get_group('honda')['price'],grouped_anova.get_group('subaru')['price'])
anova_results_1
```

```
#ANOVA test between Honda and Jaguar cars
anova_results_2 = stats.f_oneway(grouped_anova.get_group('honda')['price'],grouped_anova.get_group('jaguar')['price'])
anova_results_2
```

###**Pearson correlation and p-value**

```
#Pearson correlation and P value between Wheel-base and Length
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['length'])
pearson_coef, p_value
```
